## Preparing a complete data analysis report on the given data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
labels = pd.read_csv("Data/labels.csv")
values = pd.read_csv("Data/values.csv")

# Merge the datasets on patient_id
df = pd.merge(values, labels, on="patient_id")
df
```

```python
df_info = df.info()
df_head = df.head()
```

```python
df_description = df.describe(include='all')
df_description
```

```python
df.isnull().sum(), df_info, df_head, df_description
```

```python
# Plot target class distribution
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Heart Disease Data", fontsize=16)

sns.countplot(data=df, x='heart_disease_present', ax=axes[0, 0]) # 1. Class balance
axes[0, 0].set_title("Target Distribution")
axes[0, 0].set_xticklabels(["No Disease (0)", "Disease (1)"])

sns.histplot(data=df, x='age', hue='heart_disease_present', kde=True, ax=axes[0, 1], palette="Set2") # 2. Age distribution by heart disease

axes[0, 1].set_title("Age Distribution by Heart Disease")

sns.boxplot(data=df, x='heart_disease_present', y='max_heart_rate_achieved', ax=axes[1, 0], palette="Set3") # 3. Max heart rate by heart disease
axes[1, 0].set_title("Max Heart Rate vs Heart Disease")
axes[1, 0].set_xticklabels(["No Disease", "Disease"])

sns.countplot(data=df, x='sex', hue='heart_disease_present', ax=axes[1, 1]) # 4. Sex distribution by heart disease
axes[1, 1].set_title("Gender vs Heart Disease")
axes[1, 1].set_xticklabels(["Female", "Male"])

plt.show()
```

## model predicting potential Heart Diseases in people using Machine Learning algorithms.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Drop ID and encode 'thal'
df_model = df.drop(columns=['patient_id'])
df_model['thal'] = LabelEncoder().fit_transform(df_model['thal'])

# Features and target
X = df_model.drop(columns=['heart_disease_present'])
y = df_model['heart_disease_present']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train models
logreg = LogisticRegression(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

logreg.fit(X_train, y_train)
rf.fit(X_train, y_train)

logreg_pred = logreg.predict(X_test)
rf_pred = rf.predict(X_test)

logreg_report = classification_report(y_test, logreg_pred, output_dict=True)
rf_report = classification_report(y_test, rf_pred, output_dict=True)

accuracy_logreg = accuracy_score(y_test, logreg_pred)
accuracy_rf = accuracy_score(y_test, rf_pred)

y_pred = rf.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)

print(accuracy_logreg), print(accuracy_rf), print(logreg_report), print(rf_report)
```
